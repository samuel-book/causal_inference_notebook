# Graphviz plot of Directed Acyclic Graph (DAG) for thrombolysis use in stroke

An example DAG showing key determinants for outcome in stroke. This includes use of thrombolysis (a *clot-busting* treatment), and the patient characteristics that influence whether thrombolysis is used.

## Set up graph

```python
from graphviz import Digraph

# Create a directed graph
g = Digraph()

# Set graph attributes
g.attr(overlap='scale')
g.attr(sep='0.25')
g.attr(layout='dot') # See https://graphviz.org/docs/layouts/

# Set node attributes
g.attr('node', shape='circle', fixedsize='true', width='1.4', height='1.4')
```

## Define nodes

```python
g.node('outcome', label='Outcome', color='black',fontcolor='Black') # fontname='times bold'
g.node('nihss', label='Stroke\nseverity', color='black', fontcolor='Black')
g.node('stroke_type', label='Stroke\ntype', color='black', fontcolor='Black')
g.node('mrs', label='Prior\ndisability', color='black', fontcolor='Black')
g.node('arrive_270', label='Arrival within\n270mins', color='black', fontcolor='Black')
g.node('age', label='Age', color='black', fontcolor='Black')
g.node('af', label='Atrial\nfibrillation', color='black', fontcolor='Black')
g.node('thrombolysis', label='Thrombolysis\n(use/time)', color='black', fontcolor='Black')
g.node('thrombectomy', label='Thrombectomy\n(use/time)', color='black', fontcolor='Black')
g.node('anticoag', label='Anticoagulation', color='grey', fontcolor='Grey')
g.node('onset', label='Not known,\nEstimated or \nPrecicely known\nonset time', color='black', fontcolor='Black')
g.node('diabetes', label='Diabetes', color='black', fontcolor='Black')
g.node('hospital', label='In-hospital\nonset', color='black', fontcolor='Black')
g.node('covid', label='COVID\ninfection +\nvacc status', color='black', fontcolor='Black')
g.node('ethnicity', label='Ethnicity', color='black', fontcolor='Black')
g.node('IMD', label='Index\nMultiple\nDeprivation', color='black', fontcolor='Black')
```

## Describe edges

```python
g.edge('nihss', 'outcome', color='Black') # style='dashed')
g.edge('mrs', 'outcome', color='Black')
g.edge('age', 'outcome', color='Black')
g.edge('af', 'outcome', color='Black')
g.edge('stroke_type', 'outcome', color='Black')
g.edge('thrombolysis', 'outcome', color='Black')
g.edge('thrombectomy', 'outcome', color='Black')
g.edge('diabetes', 'outcome', color='Black')
g.edge('hospital', 'outcome', color='Black')
g.edge('covid', 'outcome', color='Black')
g.edge('ethnicity', 'outcome', color='Black')
g.edge('IMD', 'outcome', color='Black')

# Treatment confounders
g.edge('nihss', 'thrombolysis', color='Black')
g.edge('mrs', 'thrombolysis', color='Black')
g.edge('age', 'thrombolysis', color='Black')
g.edge('onset', 'thrombolysis', color='Black')
g.edge('stroke_type', 'thrombolysis', color='Black')
g.edge('anticoag', 'thrombolysis', color='Black')
g.edge('arrive_270', 'thrombolysis', color='Black')
g.edge('hospital', 'thrombolysis', color='Black')

# Treatment confounders
g.edge('nihss', 'thrombectomy', color='Black')
g.edge('mrs', 'thrombectomy', color='Black')
g.edge('age', 'thrombectomy', color='Black')
g.edge('onset', 'thrombectomy', color='Black')
g.edge('stroke_type', 'thrombectomy', color='Black')
g.edge('arrive_270', 'thrombectomy', color='Black')
g.edge('hospital', 'thrombectomy', color='Black')

# Others
g.edge('age', 'mrs', color='Black')
g.edge('af', 'anticoag', color='Black')


```

## Causal model

```python
# Create a directed graph
g = Digraph()

# Set graph attributes
g.attr(overlap='scale')
g.attr(sep='0.25')
g.attr(layout='dot') # See https://graphviz.org/docs/layouts/

# Set node attributes
g.attr('node', shape='circle', fixedsize='true', width='1.4', height='1.4')
```

```python
g.node('outcome', label='Outcome', penwidth='2', color='red',fontname='times bold', fontcolor='Red') 
g.node('nihss', label='Stroke\nseverity', color='black', fontcolor='Black')
g.node('mrs', label='Prior\ndisability', color='black', fontcolor='Black')
g.node('age', label='Age', color='black', fontcolor='Black')
g.node('af', label='Atrial\nfibrillation', color='black', fontcolor='Black')
g.node('thrombolysis', label='Thrombolysis\n(use/time)', penwidth='2', color='red', fontname='times bold', fontcolor='Red')
g.node('thrombectomy', label='Thrombectomy\n(use/time)', penwidth='2', color='red', fontname='times bold', fontcolor='Red')
g.node('onset', label='Estimated or \nPrecicely known\nonset time', color='black', fontcolor='Black')

g.edge('nihss', 'outcome', color='Black')
g.edge('mrs', 'outcome', color='Black')
g.edge('age', 'outcome', color='Black')
g.edge('onset', 'outcome', color='Black')
g.edge('af', 'outcome', color='Black')
g.edge('thrombolysis', 'outcome', color='Red')
g.edge('thrombectomy', 'outcome', color='Red')

g.edge('nihss', 'thrombolysis', color='Black')
g.edge('mrs', 'thrombolysis', color='Black')
g.edge('age', 'thrombolysis', color='Black')
g.edge('onset', 'thrombolysis', color='Black')
g.edge('af', 'thrombolysis', color='Black')

g.edge('nihss', 'thrombectomy', color='Black')
g.edge('mrs', 'thrombectomy', color='Black')
g.edge('age', 'thrombectomy', color='Black')
g.edge('onset', 'thrombectomy', color='Black')

g.edge('age', 'mrs', color='Black')

with g.subgraph() as s:
    s.attr(rank='same')
    s.node('nihss')
    s.node('age')
    s.node('mrs')
    s.node('onset')
    s.node('af')


g
```
