# Examples of causal relationships

```python
from graphviz import Digraph
```

## Causation

```python
g = Digraph(graph_attr={'rankdir': 'LR'}) # Left to right
g.attr(overlap='scale'); g.attr(sep='0.25'); g.attr(layout='dot')
g.node('x', label='x', fontcolor='Black')
g.node('y', label='y', fontcolor='Black')
# Add space to the edge label to avoid clipping

g.edge('x', 'y', label='', fontcolor='Black')
g
```

## Confounder

```python
g = Digraph(graph_attr={'rankdir': 'LR'})
g.attr(overlap='scale'); g.attr(sep='0.25'); g.attr(layout='dot')
g.node('x', label='x', fontcolor='Black')
g.node('y', label='y', fontcolor='Black')
g.node('c', label='Confounder', fontcolor='Black')
# Add space to the edge label to avoid clipping

g.edge('x', 'y', label='', fontcolor='Black')
g.edge('c', 'x', label='', fontcolor='Black')
g.edge('c', 'y', label='', fontcolor='Black')
g
```

## Mediation

```python
g = Digraph(graph_attr={'rankdir': 'LR'})
g.attr(overlap='scale'); g.attr(sep='0.25'); g.attr(layout='dot')
g.node('x', label='x', fontcolor='Black')
g.node('y', label='y', fontcolor='Black')
g.node('z', label='Mediator', fontcolor='Black')
# Add space to the edge label to avoid clipping

g.edge('x', 'z', label='', fontcolor='Black')
g.edge('z', 'y', label='', fontcolor='Black')
g
```

## Modification/interaction

```python
g = Digraph(graph_attr={'rankdir': 'LR'})
g.attr(overlap='scale'); g.attr(sep='0.25'); g.attr(layout='dot')
g.node('x1', label='x1', fontcolor='Black')
g.node('x2', label='x2', fontcolor='Black')
g.node('y', label='Y', fontcolor='Black')
g.node('i', label='Interaction', fontcolor='Black')
# Add space to the edge label to avoid clipping

g.edge('x1', 'i', label='', fontcolor='Black')
g.edge('x2', 'i', label='', fontcolor='Black')
g.edge('i', 'y', label='', fontcolor='Black')
g
```

## Collider

```python
g = Digraph(graph_attr={'rankdir': 'LR'})
g.attr(overlap='scale'); g.attr(sep='0.25'); g.attr(layout='dot')
g.node('x1', label='x1', fontcolor='Black')
g.node('x2', label='x2', fontcolor='Black')
g.node('y', label='y', fontcolor='Black')


g.edge('x1', 'y', label='', fontcolor='Black')
g.edge('x2', 'y', label='', fontcolor='Black')
g
```
